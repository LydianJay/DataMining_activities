```python
#Activity 3 Clustering 
import numpy as np
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("//home//lydianjay//Documents//Data mining//dataset//penguins.csv")
dataset.replace('NA', np.nan, inplace = True)
dataset.replace(' ', np.nan, inplace = True)
dataset.dropna( inplace = True, how = 'any')
dataset = dataset[dataset.sex != '.']


X = dataset.loc[:, ['flipper_length_mm', 'body_mass_g']].values
```

```python
dataset.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X)

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative_labels = agglomerative.fit_predict(X)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.xlabel("flipper length")
plt.ylabel("body mass")
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels)
plt.title("K-Means Clusters")
plt.xlabel("flipper length")
plt.ylabel("body mass")
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels)
plt.title("DBSCAN Clusters")
plt.xlabel("flipper length")
plt.ylabel("body mass")
plt.show()
```
